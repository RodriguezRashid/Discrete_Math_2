#### Task 1:

**Question 1:**

To find the expected value, $E[y_i]$, we need to calculate $\sum y_i \cdot P(y_i)$, which is the sum of all possible values of $y_i$ multiplied by their respective probabilities.

Given the joint probability distribution function:

$$
P(y_1, y_2, \ldots, y_n) = \frac{n!}{y_1!y_2! \cdots y_n!} p_1^{y_1} p_2^{y_2} \cdots p_n^{y_n}
$$

We want to find the expected value of $y_i$, the number of trials for which the outcome falls into cell $i$.

$$
E[y_i] = \sum_{y_1 + y_2 + \cdots + y_n = n} y_i \cdot P(y_1, y_2, \ldots, y_n)
$$

Using the linearity of expectation, we can separate the summation involving $y_i$ from the others. We know that:

$$
E[y_i] = n \cdot p_i
$$

This can be derived more formally from the properties of multinomial distributions, but the key point is that the expected number of trials falling into a specific cell $i$ is simply the total number of trials $n$ multiplied by the probability $p_i$ of falling into that cell.

### Question 1: Deriving the Binomial Probability Distribution
The binomial distribution represents the number of successes in $n$ independent Bernoulli trials. The probability of $k$ successes in $n$ trials with success probability $p$ is given by:

$$
P(Y = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

Where:
- $Y$ is the random variable representing the number of successes.
- $\binom{n}{k} = \frac{n!}{k!(n - k)!}$ is the binomial coefficient.
- $p$ is the probability of success on a single trial.
- $1 - p$ is the probability of failure on a single trial.

### Question 2: Expectation of the Binomial Distribution

$$
E[Y] = \sum_{k=0}^{n} k \cdot P(Y = k)
$$

Using properties of binomial coefficients and simplifications, it can be shown that:

$$
E[Y] = n \cdot p
$$

**Question 1:** Find a general formula for the expected value of this distribution.

**Answer 1:**

For a Poisson-distributed random variable $Y$ with parameter $\lambda$, the expected value (mean) is given by:

$$ E[Y] = \lambda $$

**Question 2:** A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.

**Answer 2:**

Given the mean density of seedlings is approximately five per square yard ($\lambda = 5$), we need to find the probability that none of the 10 regions will contain seedlings.

Let's denote the number of seedlings in a single 1-square-yard region as $Y$. For $Y$ following a Poisson distribution with $\lambda = 5$:

$$ P(Y = 0) = \frac{5^0}{0!} e^{-5} = e^{-5} $$

Since the forester examines 10 independent regions, we can extend this to 10 regions. Let $X$ be the number of regions out of the 10 that contain no seedlings. Each region is independently Poisson-distributed with $\lambda = 5$.

The probability that a single region contains no seedlings is $e^{-5}$. The probability that all 10 regions contain no seedlings is:

$$ P(\text{all 10 regions contain no seedlings}) = (e^{-5})^{10} = e^{-50} $$

### Conclusion

The probability that none of the ten 1-square-yard sampling regions will contain seedlings is $e^{-50}$.
